Una vez hecho su mayor esfuerzo, si no sabe cómo seguir matemáticamente, puede sustituir con valores y mostrar por contra-ejemplo.

a)  
$g\left(t\right)=\log _{10} u(t)$







Homogeneidad:
$$
L\left\{\alpha*g\left(t\right)\right\}=\alpha * L\left\{ u(t)\right\}
$$
Dado que 
$$
\log _{10} (\alpha*u(t)) = \log _{10} (\alpha) + \log _{10} (u(t)) ≠ \alpha*\log _{10} u(t) 
$$
no son lineales para valores de $\alpha$ diferentes a 1

$$\text{Demostración completa:}$$

$$g\left(t\right)=\log _{10} u(t)$$
$$L\left\{ \alpha f_{1}\left(x\right)+\beta f_{2}\left(x\right)\right\} =\alpha L\left\{ f_{1}(x)\right\} +\beta L\left\{ f_{2}(x)\right\}$$
$$log _{10} [\alpha u_1(t)+\beta u_2(t)] =? \alpha log _{10}u_1(t) + \beta log _{10}u_2(t)$$
$$log _{10} [\alpha u_1(t)+\beta u_2(t)] =?  log _{10}[u_1(t)]^{\alpha} + log _{10}[u_2(t)]^{\beta}$$
$$log _{10} [\alpha u_1(t)+\beta u_2(t)]  \neq  log _{10}([u_1(t)]^{\alpha} [u_2(t)]^{\beta})$$
$$ \Rightarrow \text{No es lineal} $$



b)  
$g\left(t\right)= 5 * u(t) + 13$

Homogeneidad:
Hay que comprobar que:
$$
L\left\{\alpha g\left(t\right)\right\}=\alpha L\left\{ u(t)\right\}
$$
Dado que
$$
\alpha((5u(t)) + 13) = \alpha(5u(t)) + \alpha 13 ≠ 5(\alpha u(t)) + 13 
$$
No son lineales

$$\text{Demostración completa:}$$
$$g\left(t\right)= 5 * u(t) + 13$$
$$L\left\{ \alpha f_{1}\left(x\right)+\beta f_{2}\left(x\right)\right\} =\alpha L\left\{ f_{1}(x)\right\} +\beta L\left\{ f_{2}(x)\right\}$$
$$5[\alpha u_1(t) + \beta u_2(t)]+13 =? \alpha[5u_1(t)+13]+\beta[5u_2(t)+13]$$
$$5\alpha u_1(t) + 5\beta u_2(t)+13 \neq 5\alpha u_1(t)+13\alpha +5\beta u_2(t)+13 \beta$$
$$ \Rightarrow \text{No es lineal} $$

**Pregunta 4 (20 puntos, 10 pts c/u)**

a) Proponga dos vectores:  $x$  e  $y$ que sean colineales (con dos elementos cada uno).  Programe en python para mostrar que son colineales y luego grafíquelos en un mismo gráfico en python. En el cuaderno visto en clase viene un ejemplo de uso de import matplotlib.pyplot as plt



```python
def colineales (array1, array2):
        #print ("x =" + np.array2string(array1))
        #print("y = " + np.array2string(array2))
 
        if array1.shape != array2.shape:
                return("Both arrays must have the same shape.")
        quotients = []
        for i, a in enumerate(array1):
                for j , b in enumerate(array2):
                        if i==j:
                                quotients.append(a/b)
        if all(x == quotients[0] for x in quotients):
                return ("Son colineales con \u03B1 = " + str(quotients[0]))
        else:
                return "No son colineales"
                           
x = np.array([ 3. , 7.])
y = np.array([9. , 21.])

colineales( x , y)

X = np.array([0.,0.] )
Y = np.array([0.,0.] )
U = np.array(x)
V = np.array(y)

fig, ax = plt.subplots()
ax.quiver(X, Y, U ,V ,color=["r","b"],angles='xy', scale_units='xy', scale=1)
ax.axis([-5.,25.,-5.,25.])
ax.set_aspect('equal')
plt.show()
```
